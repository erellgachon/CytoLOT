```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bary_kmeans_times = np.load("./Results/Exec_times/bary_kmeans_times.npy")
bary_flowsom_times = np.load("./Results/Exec_times/bary_flowsom_times.npy")
linw2_kmeans_times = np.load("./Results/Exec_times/linW2_kmeans_times.npy")
linw2_flowsom_times = np.load("./Results/Exec_times/linW2_flowsom_times.npy")
comp_times = np.load("./Results/Exec_times/lR-times.npy")
kme_times = np.load("./Results/Exec_times/KME_times.npy")
flowsom_times = np.load("./Results/Exec_times/flowsom_times.npy")
kmeans_times = np.load("./Results/Exec_times/KMeans_times.npy")
```

```python
kme_times = [kme_times[0], kme_times[0], kme_times[0], kme_times[0], kme_times[0]]
```

```python
nb_clusters = np.array([16,32,64,128,256])
```

```python
fig = plt.figure(figsize=(13,8))
plt.plot(nb_clusters, (kmeans_times+bary_kmeans_times+linw2_kmeans_times)[:-1], marker='o', c='blue', label="K-means + LinW2 (ref=barycenter)")
plt.plot(nb_clusters, (kmeans_times+linw2_kmeans_times)[:-1],marker='o', c='purple', label="KMeans + LinW2 (ref=uniform)")
plt.plot(nb_clusters, (flowsom_times+bary_flowsom_times+linw2_flowsom_times)[:-1], marker='o', c='orange', label="FlowSOM + LinW2 (ref=barycenter)")
plt.plot(nb_clusters, (flowsom_times+linw2_flowsom_times)[:-1], marker='o', c='red', label="FlowSOM + LinW2 (ref=uniform)")
plt.plot(nb_clusters, (kme_times), marker='o', c='green', label="KME")
plt.xticks(nb_clusters)
#plt.xlim([0,300])
#plt.ylim([0,1000])
plt.legend(prop={'size': 15})
plt.xlabel("Number of clusters", fontsize=20)
plt.ylabel("Execution time (in seconds)", fontsize=20)
plt.savefig("./Results/images/execution_times.png")
```
